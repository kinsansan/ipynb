```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
```

```python
ary = np.random.rand(1000)
ary *= 10
ary = ary.astype(int)
```

```python
ary[:10]
```

```python
count = Counter(ary)
ranked_data = count.most_common()
```

```python
labels = [x[0] for x in ranked_data]
values = [x[1] for x in ranked_data]
```

```python
plt.pie(values)
plt.show()
```

```python
plt.pie(values)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(values)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(5,5), dpi=100)
plt.pie(values)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(5,5), dpi=100)
plt.pie(values, labels=labels)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(5,5), dpi=100)
plt.pie(values, labels=labels, startangle=90)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(5,5), dpi=100)
plt.pie(values, labels=labels, startangle=90, counterclock=False)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(5,5), dpi=100)
plt.pie(values, labels=labels, startangle=90, counterclock=False, autopct='%.1f%%')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(5,5), dpi=100)
plt.title('pie')
plt.pie(values, labels=labels, startangle=90, counterclock=False, autopct='%.1f%%')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(5,5), dpi=100)
plt.title('pie')
plt.pie(values, startangle=90, counterclock=False, autopct='%.1f%%')
plt.legend(labels)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(5,5), dpi=100)
plt.title('pie')
plt.pie(values, startangle=90, counterclock=False, autopct='%.1f%%')
plt.legend(labels, bbox_to_anchor=(0.95, 0.8))
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(5,5), dpi=100)
plt.title('pie')
plt.pie(values, startangle=90, counterclock=False, autopct='%.1f%%', pctdistance=1.15)
plt.subplots_adjust(left=0, right=1.2)
plt.legend(labels,  bbox_to_anchor=(0.98, 0.7))
plt.axis('equal')
plt.savefig("pie.png",bbox_inches='tight')
plt.show()
```
